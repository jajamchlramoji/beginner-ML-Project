## Data Cleaning

```python
teams[teams.isnull().any(axis=1)]
```

```python
teams = teams.dropna()
```

```python
teams
```

```python
teams[teams.isnull().any(axis=1)]
```

Its shows no Null values after dropping them

Split the data to model.

```python
train = teams[teams["year"]<2012].copy()
test = teams[teams["year"]>=2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
# Train the model 
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
predictors = ["athletes","prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictors],train["medals"])
```

We will now use test data with the predictors to see whether our model predicts the data that it hasn't seen before 

```python
predictions= reg.predict(test[predictors])
```

```python
predictions
```

we got a numpy array. these aren't rounded. we need medals in whole number.

Also 2nd ly that countries can't win negative medals. 

```python
#lets first assign raw predictions 
# -- to test data frame as a new column
test["predictions"]=predictions
```

```python
test
```

```python
#this will change the predictions column with <0 to zero

test.loc[test["predictions"]<0,"predictions"]=0
```

```python
#this will change the predictions we have in decimals to nearest whole number
test["predictions"] = test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test["medals"],test["predictions"])
```

```python
error
```

On average, our predictions were off by about 3.3 medals compared to the actual number of medals a team won in the Olympics. This value shows how close or far our model's predictions are from the real results.

```python
#lets find whthere good or wrong error 

teams.describe()["medals"]
```

Here if the error value is below std deviation its nice but  if its higher then usaullly sth is wrong. like data or predictors and so on.

```python
errors=(test["medals"]- test["predictions"]).abs()

#MEAN ABSOLUTE ERROR
```

```python
errors
```

```python
#now we wil group by team and show how much we were off per TEAM. mean shows meanerror of each team -> test["team"]

error_by_team=errors.groupby(test["team"]).mean()
```

```python
error_by_team
```

```python
medals_by_team= test["medals"].groupby(test["team"]).mean()
```

```python
error_ratio = error_by_team/medals_by_team
```

```python
error_ratio
```

```python
### WE have a lot of NaN values. so now we clean!!! remove nAn 
error_ratio[~pd.isnull(error_ratio)]
```

```python
#we remove inf values too 
import numpy as np
error_ratio = error_ratio[np.isfinite(error_ratio)]
```
